```python
import json
import pprint
from copy import deepcopy
```

```python
with open('/content/drive/MyDrive/QA/train-v2-back-translation-aug.json',encoding='utf-8') as f:
    data = json.load(f)
```

```python
uniques = []
data_tmp = deepcopy(data)
for i in range(len(data_tmp['data'])):
  for j in range(len(data_tmp['data'][i]['paragraphs'])):
    for k in reversed(range(len(data_tmp['data'][i]['paragraphs'][j]['qas']))):
      if data_tmp['data'][i]['paragraphs'][j]['qas'][k]['id'] not in uniques:
        uniques.append(str(data_tmp['data'][i]['paragraphs'][j]['qas'][k]['id']))
        print('dropped')
        data['data'][i]['paragraphs'][j]['qas'].pop(k)
```

```python
 with open('/content/drive/MyDrive/QA/train-v2.0-back-aug.json', 'w') as outfile:
    json.dump(data, outfile)
```
