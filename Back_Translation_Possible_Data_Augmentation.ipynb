```python
import json
```

```python
with open('/content/drive/MyDrive/QA/train-v2.0.json',encoding='utf-8') as f:
    data = json.load(f)
```

```python
with open('/content/drive/MyDrive/QA/train-v2.0-back-aug.json',encoding='utf-8') as f:
    aug_data = json.load(f)
```

```python
for i in range(0,len(aug_data["data"])):
    for j in range(0,len(aug_data["data"][i]["paragraphs"])):
        for k in range(0,len(aug_data["data"][i]["paragraphs"][j]["qas"])):
          if aug_data["data"][i]["paragraphs"][j]["qas"][k]['is_impossible'] == 'false':
            aug_data["data"][i]["paragraphs"][j]["qas"][k] = data["data"][i]["paragraphs"][j]["qas"][k]

with open('train-v2-possible-back-trans-aug.json', 'w') as outfile:
    json.dump(aug_data, outfile)    
```
